```python
from square import square # import the square function we wrote
import numpy as np
```

```python

# use the range function to generate the integers 1 through 10
x = range(1,11)
xs = np.array(x)
# for n in x:
#     print(n)
# xs = ### YOUR CODE HERE ###
# square each element in xs
# xs_squared = ### YOUR CODE HERE ###
xs_squared = square(xs)
print("List of x values:", xs)
print("List of x^2 values:", xs_squared)
```

```python
import matplotlib.pyplot as plt
plt.scatter(xs,xs_squared, label = 'y = x^2')
plt.title("Scatter plot: y = x^2")
plt.xlabel("X")
plt.ylabel("X^2")
plt.show()
```
